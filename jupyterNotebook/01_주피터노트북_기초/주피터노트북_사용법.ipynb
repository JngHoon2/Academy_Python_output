# 주피터 노트북 사용법    
#1. 셀이 초록색 : 편집모드 - 셀을 클릭하면 코딩을 할 수 있는 상태로 바뀜.
  - 편집모드에서 [Esc]를 누르면 명령모드로 바뀜.  
  - 편집모드에서 [b]를 누르면 바로 밑에 셀이 추가된다.  
  
#2. 셀이 파란색 : 명령모드
  - 명령모드에서 [M]을 누르면 Markdown 모드로 바뀜.  
  - 마크다운으로 바뀌면 in[숫자] 부분이 사라짐. 
  
#3. 셀 실행 : [ctrl] + [Enter] : 실행 후 현재 셀에 위치  
#4. 셀 실행 : [shift] + [Enter] : 실행 후 다음 라인에 새로운 셀 생성  
#5. 셀 추가 : 명령모드에서 [insert + insert cell above]  
#6. 셀별 라인 넘버 : [View > toggle line number]
