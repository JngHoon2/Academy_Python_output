# 그래프 그리기
#### 데이터 시각화 도구 matplotlib 사용하기

 - matplotlib 라이브러리 import

```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

pylab.rcParams['font.size'] = 8
pylab.rcParams['figure.figsize'] = (8, 6)

plt.rcParams['axes.unicode_minus'] = False
```

 - pyplot.plot() 함수에 하나의 숫자 리스트를 입력
 - plot() 함수는 리스트의 값들이 y 값이라고 가정, x 값[0,1,2,3]을 자동으로 생성
 - show() 함수는 그래프를 화면에 나타나도록 한다.

```python
plt.plot([1,2,3,4])
plt.show()
```

- 임의의 개수의 인자를 받을 수 있다. 
- 두개의 리스트를 전달하면 x-y 값을 그래프로 나타낼 수 있다.

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
# (1,1), (2,4), (3,9), (4,16)
```

- 선의 생상과 형태를 지정하는 포맷 문자열을 세번째 인자로 전달
- 포맷 문자열 'ro'는 빨간색('red')의 원형('o') 마커를 의미
- axis() 함수를 이용하여 축의 범위[xmin, xmax, ymin, ymax]를 지정

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20]) #[xmin, xmax, ymin, ymax]
plt.show()
```

- 파이썬 딕셔너리와 같이 레이블이 있는 데이터를 그래프로 나타낼 수 있다.
- 먼저 plot() 함수에 데이터의 레이블(딕셔너리의 키)을 입력해주고,  
  data 파라미터애 딕셔너리를 지정

```python
data_dict = {'data_x' : [1,2,3,4,5], 'data_y' : [2,3,5,10,8]}

plt.plot('data_x', 'data_y', data = data_dict)
plt.show()
```

 - xlabel(), ylabel() 함수를 사용하면 그래프의 x,y 축에 대한 레이블을 표시

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-Lable')
plt.ylabel('Y-Lable')
plt.show()
```

- 폰트 설정하기

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis', labelpad = 15, fontdict = font1)
plt.ylabel('Y-Axis', labelpad = 20, fontdict = font2)
plt.show()
```

위치 지정하기

- xlabel() 함수의 loc 파라미터는 X축 레이블의 위치를 지정합니다.({'left', 'center', ''right})
- ylabel() 함수의 loc 파라미터는 Y축 레이블의 위치를 지정합니다.({'bottom', 'center', ''top})

```python
plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis', loc = 'right')
plt.ylabel('Y-Axis', loc = 'top')
plt.show()
```

범례(legend) 표시하기

 - 영역에 범례를 나타내기 위해서는 우선 plot() 함수에 label 문자열을 지정하고,
 - matplotlib.pyplot 모듈의 legend() 함수를 호출

```python
plt.plot([1,2,3,4], [2,3,5,10], label='Price($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

plt.show()
```

```python
plt.plot([1,2,3,4], [2,3,5,10], label='Price($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right') # 범례의 위치 지정 

plt.show()
```

```python
plt.plot([1,2,3,4], [2,3,5,10], label='Price($)')
plt.plot([1,2,3,4], [3,5,9,7], label='Demend(#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend(loc='best') # ncol = 1 이면 범례가 세로로 나열 
#plt.legend(loc='best', ncol=2) # ncol = 2 이면 범례가 가로로 나열

plt.show()
```

# 8.2 다양한 유형의 그래프 그리기
### 8.2.1 막대그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

plt.title("막대그래프(bar)")
x_data = [1,3,5,7,9]
y_data = [5,7,6,1,4]

plt.bar(x_data, y_data) #bar(막대를 표시할 위치, 막대의 높이-y값)
plt.show()
```

```python
x = np.arange(3) # [0, 1, 2] # 주어진 범위와 간격에 따라 균일한 값을 갖는 배열을 반환
x_years = ['2018', '2019', '2020'] # x_years X축에 표시될 연도
y_values = [100, 400, 900] # y_values 막대 그래프의 y값

#bar() 함수에 0부터 시작하는 정수 리스트 넣어줌
plt.bar(x, y_values) # bar() 함수에 x값 [0,1,2]와 y값 [100, 400, 900]을 입력

# x와 x_years 입력해주면, x축의 눈금 레이블에 '2018', '2019', '2020' 순서대로 표시
# xticks() 함수에 x와 years를 입력해주면, x축의 눈금 레이블에 '2018', '2019', '2020'순서대로 표기
plt.xticks(x, x_years)
plt.show()
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, color = colors)
plt.xticks(x, years)
plt.show()
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.barh(x, values, color = colors)
plt.yticks(x, years)
plt.show()
```

### 원그래프 그리기

```python
b_type = [25, 19, 37, 11]
plt.pie(b_type, labels = ['A형', 'B형', 'O형', 'AB형'])
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]

plt.pie(ratio, labels=labels, autopct = '%.lf%%', startangle = 260,
        counterclock = False, explode=explode)

plt.legend()
plt.axis('equal')
plt.show()
```
