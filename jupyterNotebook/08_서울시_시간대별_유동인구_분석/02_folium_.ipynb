```python
import csv
```

```python
# 인구 데이터 읽어와서 리스트로 저장하기

f1 = open('./seoul_dong_people_201912.csv', encoding='utf8')
data = csv.reader(f1)
next(data)
data = list(data) # 읽어들인 파일을 2차원 리스트로 변환

print('data 건수 : ', len(data))
print('data[0] :', data[1])
```

```python
# 행정동코드 데이터 파일 읽기
f2 = open('./dong_code.csv', encoding='cp949')
code_data = csv.reader(f2)

next(code_data)
next(code_data)

code_data = list(code_data)
len(code_data)
```

```python
# 위치 데이터 불러와서 리스트로 저장하기
f3 = open('./dong_location.csv', encoding='cp949')
location_data = csv.reader(f3)
next(location_data)
location_data = list(location_data)
print(len(code_data))
```

```python
# 인구데이터 숫자형으로 변환하기
for row in data :
    for i in range(1, 32):
        if i <= 2: # 인덱스가 2이하인 경우(시간대구분, 행정동코드 등)
            row[i] = int(row[i]) # 정수형으로 변환
        else : # 인덱스다 2초과인 경우(총생활인구수부터)
            row[i] = float(row[i]) # 실수형으로 변환
            
print(data[0])
```

```python
for row in code_data :
    row[0], row[1] = int(row[0]), int(row[1])
    
for row in location_data :
    row[2], row[-2], row[-1] = int(row[2]), float(row[-2]), float(row[-1])
```

```python
# 동별 인구수를 저장하는 딕셔너리 생성하고 동별 총생활인구를 딕셔너리에 저장
# {동코드1 : 총생활인구수1, 동코드2 : 총생활인구수2, ....} 형태로 데이터 저장
dong_population = {}

for row in data :
    dong_code, p = row[2], row[3] # 동코드, 총생활인구
    if dong_code not in dong_population.keys():
        dong_population[dong_code] = p # 아직 키가 없으면 키-값 추가
    else:
        dong_population[dong_code] += p # 이미 키가 만들어졌으면 합산

dong_population
```

```python
# 인구수를 기준으로 내림차순 정렬하고, 앞에서 10개 항목만 별도로 저장하기
import operator

#{동코드:인구수} operator.itemgetter(1): 인구수를 말함.
# top10 type : list
top10 = sorted(dong_population.items(), key=operator.itemgetter(1), reverse=True)[:10]
print(top10)

top10_code8 = [t[0] for t in top10] # 리스트 컴프리헨션으로 동코드만 선별
print(top10_code8)
```

```python
# 동별 위도, 경도 정보 저장용 변수
top10_name = [0 for i in range(10)]
top10_lat = [0 for i in range(10)]
top10_lng = [0 for i in range(10)]

# 8자리 동코드에 해당하는 동명 찾기
for i in range(len(top10_code8)):
    for row in code_data:
        if row[1] == top10_code8[i]:
            # 행정동명 리스트에 저장하기
            top10_name[i] = row[-1]
    # 행정동명에 해당하는 위도와 경도를 찾아 리스트에 저장하기
    for row in location_data:
        if row[3] == top10_name[i]:
            top10_lat[i], top10_lng[i] = row[-1], row[-2]

print(top10_lat)
print(top10_lng)
    
```

```python
# 추가. 헹정동명으러 위치 정보를 찾지 못하면 7자리 행정동코드로 찾기
for i in range(len(top10_lat)):
    # 위도나 경도가 0인 요소가 존재하면
    if top10_lat[i] == 0 or top10_lng[i] == 0:
        # code_data에서 7자리 행정동코드를 구해 dong7에 저장하기
        for row in code_data :
            if top10_code8[i] == row[1]:
                dong7 = row[0]
        # location_data에서 7자리 행정동코드가 dong7인 행을 찾아 위도와 경도를 구함.
        for row in location_data:
            if dong7 == row[2]:
                top10_lat[i], top10_lng[i] = row[-1], row[-2]
print(top10_lat)
print(top10_lng)
```

```python
import folium

for i in range(len(top10_name)) :
    top10_name[i] = str(i+1) + '. ' + top10_name[i]
    
map_top10 = folium.Map([top10_lat[0], top10_lng[0]], zoom_start=12)

for i in range(10):
    folium.Marker(
        [top10_lat[i], top10_lng[i]], tooltip=top10_name[i],
         icon=folium.Icon(color='red', icon='star')
    ).add_to(map_top10)
    
map_top10
```
