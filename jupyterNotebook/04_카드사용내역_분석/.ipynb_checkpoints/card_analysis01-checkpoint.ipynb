# 데이터 준비 : 파일에서 데이터 읽어들이기
#### CSV 파일이란?
#### CSV 파일 읽어들이기

- CSV 파일을 다루려면 CSV 파이브러리가 필요합니다. import csv
- 한 주피터 노트북 파일에서는 한 번만 import 하면 셀마다 할 필요는 없습니다.

카드사용내역 csv 파일을 읽어들여서 데이터분석
 - 파일 read/write 시 인코딩은 가급적이면 저장시 사용했던 인코딩 방식으로 read할 것!
 - 공공데이터셋(전자정부프레임워크)은 CP949로 지정

```python
import pandas as pd
import os
import csv

# csv 파일이 위치한 경로(파일명)
filePath = './card.csv'

# 작성중인 노트북의 현재 위치 확인
# print(os.getcwd())

# card.csv파일 읽기
df = pd.read_csv(filePath, encoding='utf-8', skiprows=0)
print(df)
```

```python
print(df.shape) # 데이터프레임의 모양
print(df.columns) # 컬럼 인덱스(컬럼명)
print(df.index) # 행 인덱스(일시)
```

 - range + iloc을 통한 순회[반복]

```python
for i in range(len(df)):
     print(df.iloc[i, 0], df.iloc[i, 1], df.iloc[i, 2], df.iloc[i, 3], df.iloc[i, 4], 
           df.iloc[i, 5], df.iloc[i, 6], df.iloc[i, 7])
```

```python
for row in df.itertuples():
    print(row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7])
```

 - 200,000 < 이용금액 < 250,000 사이인 거래내역만 조회
 1. 조건을 데이터프레임에 바로 적용하는 경우

```python
print(df[(df['이용금액'] >= 200000) & (df['이용금액'] < 250000)])
```

- 200,000 < 이용금액 < 250,000 사이인 거래내역만 조회
 2. 조건을 별도로 만들어서 적용하는 경우

```python
# mask1의 type은 Series로 True, False, True...... 형태의 값을 갖는다.
# 특정 컬럼에 적용할 필터링 조건
mask1 = (df['이용금액'] >= 200000) & (df['이용금액'] < 250000) 

# 행을 선별하는데 mask1의 True/False 값이 사용된다.
# mask1 필터링 조건에 해당하는 행이 선별되어 새로운 dataframe에 저장됨.  
df_sum_condition = df.loc[mask1, :] 
print(df_sum_condition)
```

 - 조건1) 50,000 < 이용금액 < 100,000 사이
 - 조건2) 가맹점명이 네이버파이낸셜(주)인 겅우의 거래내역만 조회

```python
mask1 = (df['이용금액'] >= 50000) & (df['이용금액'] < 100000) 

mask2 = (df['가맹점명'] == '네이버파이낸셜(주)')

df_sum_condition = df.loc[mask1 & mask2, :] 
print(df_sum_condition)
```

```python
# 10월에 결제한 내역만 조회
df4 = df[df.apply(lambda x : x['이용일시'].split()[0].split('/')[1] == '10', axis = 1)]
print(df4)
```
