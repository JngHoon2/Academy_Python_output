# 시카고 샌드위치 맛집 소개 사이트 Crawling

- 이제 시카고의 베스트 샌드위치 가게를 소개하고 있는 시카고 매거진  
  홈페이지에 접속하여 리스팅 된 샌드위치 가게 장보를 얻어오자
  
  https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/
  
  

> ## 목표
1. 메뉴의 이름과 가게의 이름이 있다. 그리고 Read More이라는 버튼을 누르면 시카고 매거진에서  
   각 가게를 리뷰한 페이지로 넘어가게 된다.  
2. 일단 목표는 가게 이름, 가게 메인 메뉴, 각 가게 소개 페이지를 정리하는 것으로 한다.

 - 필요한 라이브러리 import

```python
from bs4 import BeautifulSoup
#from urllib.request import Request, urlopen
import requests
```

 - 기본 url 및 헤더 설정 후 파싱

```python
headers = {'User-Agent': 'Mozilla/5.0'}

url = 'http://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'

html = requests.get(url, headers = headers).text

soup = BeautifulSoup(html, "html.parser")
soup
```

```python
# find_all 명령을 이용해서 div의 sammy 태그를 찾는다.
print(soup.find_all('div', 'sammy')[:3])
```
