> ## 서울시 구별 인구현황(엑셀파일)

```python
# 서울 구별 인구 raw 데이터 위치
# 서울 구별 인구 dataframe
filePath = '01_population_in_seoul.xls'

# excel read, header=2(3번 째 줄부터 읽어라)
# B, D, G, J, N 컬럼만 읽어옴
pop_Seoul = pd.read_excel(filePath, header = 2, usecols = 'B, D, G, J, N')
pop_Seoul.head()
```

 - 데이터프레임의 컬럼명 변경

> ## 최근 CCTV 증가율(%) 계산

 - 2014년부터 2016년까지 3년간 CCTV수를 더하고 2013년 이전 CCTV 수로 나누어 최근 3년간 CCTV증가율 

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2016년'])
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
CCTV_Seoul['증가율'] = round(CCTV_Seoul['최근증가율'] / CCTV_Seoul['2013년도 이전'] * 100, 2)
CCTV_Seoul.sort_values(by='증가율', ascending=False).head()
```

  > ## merge() 함수로 두 데이터프레임 합치기

```python
left = pd.DataFrame({'key' : ['K0', 'K4', 'K2', 'K3'],
                     'A' : ['A0', 'A1', 'A2', 'A3'],
                     'B' : ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key' : ['K0', 'K1', 'K2', 'K3'],
                     'C' : ['C0', 'C1', 'C2', 'C3'],
                     'D' : ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

 - 두 데이터에 공통으로 있는 컬럼을 기준으로 merge명령에서 merge 기준을 설정하는  
   on옵션으로 합치면 공통된 key에 대해서만 합치게 됨.

```python
pd.merge(left, right, on='key')
```

```python
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

```python
pd.merge(left, right, how='outer', on='key')
```

> ## CCTV 데이터와 인구 데이터 합치고 분석하기

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['최근증가율']
data_result.head()
```
