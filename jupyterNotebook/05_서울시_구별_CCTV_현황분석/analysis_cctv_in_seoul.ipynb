> ## 서울시 구별 인구현황(엑셀파일)

```python
# 서울 구별 인구 raw 데이터 위치
# 서울 구별 인구 dataframe
filePath = '01_population_in_seoul.xls'

# excel read, header=2(3번 째 줄부터 읽어라)
# B, D, G, J, N 컬럼만 읽어옴
pop_Seoul = pd.read_excel(filePath, header = 2, usecols = 'B, D, G, J, N')
pop_Seoul.head()
```

 - 데이터프레임의 컬럼명 변경

```python
pop_Seoul.rename(columns = { pop_Seoul.columns[0] : '구별',
                             pop_Seoul.columns[1] : '인구수',
                             pop_Seoul.columns[2] : '한국인',
                             pop_Seoul.columns[3] : '외국인',
                             pop_Seoul.columns[4] : '고령자'}, inplace = True)
pop_Seoul.head()
```

 - 소계로 오름차순 정렬

> ## 최근 CCTV 증가율(%) 계산

 - 2014년부터 2016년까지 3년간 CCTV수를 더하고 2013년 이전 CCTV 수로 나누어 최근 3년간 CCTV증가율 

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2016년'])
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
CCTV_Seoul['증가율'] = round(CCTV_Seoul['최근증가율'] / CCTV_Seoul['2013년도 이전'] * 100, 2)
CCTV_Seoul.sort_values(by='증가율', ascending=False).head()
```

 - 각 구별로 전체 인구대비 외국인과 고령자 비율 계산

```python
pop_Seoul['외국인비율'] = round(pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100, 2)
pop_Seoul['고령자비율'] = round(pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100, 2)
pop_Seoul.head()
```
