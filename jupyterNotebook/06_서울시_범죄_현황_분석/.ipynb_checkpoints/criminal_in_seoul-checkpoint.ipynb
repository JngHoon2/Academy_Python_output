> ### 경찰서 명칭 변경
- 경찰서명을 정식명칭으로 바꿈
- 정식 명칭으로 구글 map api에서 정식 주소와 위경도 정보 추출

```python
## 관서명 데이터 정리
station_name=[] # 정리된 관서명 저장
 
# 관서명을 '중부서' -> '중부경찰서'로 바꿈
for name in crime_anal_police['관서명']:
    station_name.append('서울' + str(name[:-1])+'경찰서')
    
### 서울시 경찰서의 주소, 위치 저장
station_address=[] # 서울시 경찰서의 주소, 구글 map 정보에서('formatted_address')항목

# 구글맵이 응답한 데이터 중, 'geometry':{'location': {'lat' : 37.5636465,
# 'lng': 126.9895796},
station_lat=[] # 'lat':37.5636465 위도
station_lng=[] # 'lng' :126.9895796 경도

# 경찰서명으로 구글에서 조회해서 주소, 위, 경도 정보 추출
for name in station_name:
    # 한 경찰서의 모든 정보가 list type으로 반환
    tmp=gmaps.geocode(name, language='ko')
    
    # 정식주소 추출, tmp : list type, tmp[0] : dict type
    station_address.append(tmp[0].get('formatted_address'))
    
    # 위.경도 정보 추출
    tmp_loc=tmp[0].get('geometry')
    
    # 위도, 경도 따로 분리
    station_lat.append(tmp_loc['location']['lat'])
    station_lat.append(tmp_loc['location']['lng'])
    
    # 바로 확인 (테스트용)
    print(name+"====> " +tmp[0].get('formatted_address'))
```

# 피벗 pivot_table() 함수를 이용하여 중복 합산
- pivot_table()을 이용하여 '관서별'=>'구별'

[사용방법] pivot_table(data, values=None, index=None, columns=None,
aggfunc='mean', fill_value=None, margins=False,
margins_name='All')
- data: 분석할 데이터프레임
- values: 테이블의 값이 될 정보
- index: row name으로 나타낼 정보가 담긴 컬럼
- columns: column name으로 나타낼 정보가 담긴 컬럼
- aggfunc: 두 데이터를 합쳐야하는데 합칠 때 어떤 연산을 적용할지를 의미, 
   
    기본은 평균 np, mean
    
    만약 조건에 따른 데이터가 유일하게 선택되지 않으면 그룹연산을 하며 이 때
   
    aggfunc 인수로 정의된 함수를 수행하여 대표값을 계산
- fill_value: 합친 후에 비어 있는 즉, NaN 값을 대체할 값을 설정
- margins: 모든 데이터를 연산한 결과를 오른쪽과 아래에 붙일지 여부
- margins_name: 마진 열(행)의 이름

- 피벗(pivot)테이블 예제#1

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns=["도시", "연도", "인구", "지역"]
df1=pd.DataFrame(data, columns=columns)
df1.head()
```

- 도시 이름이 열 인덱스가 되고 연도가 행 인덱스가 되도록 만들어보자.
- 피봇 테이블의 행인덱스 인수로는 "도시", 열 인덱스 인수로는 "연도",
    
    데이터 이름으로 "인구"를 입력하면 된다.

```python
pd.pivot_table(df1, index="도시", columns="연도", values="인구", aggfunc=np.mean)
```

```python
df1.pivot("도시", "연도", "인구")
```

- 해당 행과 해당 열, 그리고 모든 행과 열의 합산 계산

```python
df1.pivot_table(values="인구", index="도시", columns="연도", aggfunc=np.sum,
                margins=True, margins_name="합계")
```

- 5대 중요 범죄에 대한 검거율을 계산하여 crime_anal에 추가
    - *[주의]

        다음 단계에서 불필요한 컬럼을 삭제하게 되는데 삭제하고 나서 아래 5대 중요범죄 검거율 계산시 오류

        그러면 위에 올라가서 다시 데이터를 만들고 와서 실행할것.


```python
# '구별' 컬럼을 인덱스로 하여 중복되는 내용의 합을 구함
crime_anal = pd.pivot_table(crime_anal_raw, index='구별', aggfunc=np.sum) 
crime_anal
```

```python
# 5대 중요 범죄에 대한 검거율을 계산하여 crime_anal에 추가 
crime_anal['강간검거율'] = round(crime_anal['강간 검거'] / crime_anal['강간 발생'] * 100, 2)
crime_anal['강도검거율'] = round(crime_anal['강도 검거'] / crime_anal['강도 발생'] * 100, 2)
crime_anal['살인검거율'] = round(crime_anal['살인 검거'] / crime_anal['살인 발생'] * 100, 2)
crime_anal['절도검거율'] = round(crime_anal['절도 검거'] / crime_anal['절도 발생'] * 100, 2)
crime_anal['폭력검거율'] = round(crime_anal['폭력 검거'] / crime_anal['폭력 발생'] * 100, 2)

crime_anal
```

```python
del crime_anal['강간 검거']
del crime_anal['강도 검거']
del crime_anal['살인 검거']
del crime_anal['절도 검거']
del crime_anal['폭력 검거']
crime_anal
```

```python
#### 데이터 스케일링 작업 : 검거율이 100을 넘는 데이터는 100으로 변경...

for row in crime_anal.index:
    crime_anal.loc[row, '강간검거율'] = 100 if crime_anal.loc[row, '강간검거율']> 100 else crime_anal.loc[row,'강간검거율']
    crime_anal.loc[row, '강도검거율'] = 100 if crime_anal.loc[row, '강도검거율']> 100 else crime_anal.loc[row,'강도검거율']
    crime_anal.loc[row, '살인검거율'] = 100 if crime_anal.loc[row, '살인검거율']> 100 else crime_anal.loc[row,'살인검거율']                                
    crime_anal.loc[row, '절도검거율'] = 100 if crime_anal.loc[row, '절도검거율']> 100 else crime_anal.loc[row,'절도검거율']   
    crime_anal.loc[row, '폭력검거율'] = 100 if crime_anal.loc[row, '폭력검거율']> 100 else crime_anal.loc[row,'폭력검거율']  
crime_anal                                                           
```

```python
crime_anal.rename(columns={'강간 발생':'강간',
                            '강도 발생': '강도',
                            '살인 발생': '살인',
                            '절도 발생': '절도',
                            '폭력 발생': '폭력'}, inplace=True)
crime_anal
```

## 데이터 스케일링

- 대부분의 학습 알고리즘은 특성 간 데이터 범위가 다를 경우 잘 동작하지 않는다.
그 이유는 다음과 같다. 예를 들어, 나이와 재산이라는 특성이 있을 때, 재산의 값이
나이보다 훨씬 크기 때문에 재산에 치중한 학습을 하게 된다. 즉 비교 자체가 무의미해짐
- 예측 모델링을 할 때 학습 데이터에 대하여 변환이 필요한데 그중에서 데이터를 특정 범위로 
제한시키는 Min Max 변환을 많이 사용한다. 최대 최소 변환은 데이터를 지정한 범위 내로
한정시키는 것을 말한다.
- Scikit-Learn에서는 Min Max 변환 클래스인 MinMaxScaler를 제공한다.
- MinMaxScaler는 모든 요소들이 0과 1사이의 데이터값을 갖도록 만들어준다. '정규화
(Normalization)'라고 한다.
- 즉, 컬럼 범위에서 최솟값은 0이 되고, 최댓값은 1이 되는 것.
- 머신러닝을 위해 데이터 전처리를 하는 과정에서 데이터 스케일링이 중요함.
- 비교할 데이터인 강도, 살인은 두 자릿수, 절도, 폭력은 4자릿수로 범위가 너무 큼
- 비교할 데이터의 두 데이터간 범위가 너무 커서 최소값과 최대값에 대한 비율을 전처리 해야 
한다.
- 각 항목의 최대값을 1로 두기로 가정하고 각 컬럼 별로 '정규화'를 진행한다.

## 사이킷런

- 사이킷런은 파이썬에서 머신러닝 분석을 할 때 데이터 전처리를 효과적으로 해주는 라이브러리이다.
- 데이터 스케일링을 위한 패키지인 sklearn preprocessing import
- preprocessing(모듈)의 MinMaxScalar 클래스 이용

- 사이킷런 인스톨

```python
from sklearn import preprocessing

# 1. 전처리할 데이터 컬럼명 생성
col=['강간', '강도', '살인', '절도', '폭력']

# 2. 각 컬럼의 데이터를 추출
x=crime_anal[col].values

# 3. MinMaxScaler 클래스의 객체 생성
min_max_scaler=preprocessing.MinMaxScaler()

# 4. MinMaxScaler 객체에 스케일링할 데이터를 전달
# 각 컬럼별로 컬럼 내에서 차지하는 포션만큼 정규화 한다.
# feature의 최솟값과 최댓값을 기준으로 0~1 구간 내에 균등하게 값을 배정
# fit-transformer API를 통해서 '0~1' 범위 내에서 최소, 최대 '0~1'사이로 변환
x_scaled=min_max_scaler.fit_transform(x.astype(float))

# 5. 정규화가 완료된 결과 데이터를 새로운 데이터프레임에 저장
crime_anal_norm=pd.DataFrame(x_scaled, columns=col, index=crime_anal.index)

# crime_anal_norm에 검거율 추가
col2=['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']

# crime_anal 데이터프레임에서 col2에 해당하는 컬럼들 복사
crime_anal_norm[col2]=crime_anal[col2]
crime_anal_norm.head()
```

```python
# CCTV 데이터 로드 : 01.CCTV_result.csv
result_CCTV = pd.read_csv("./01.CCTV_result.csv", encoding='cp949', index_col='구별')
# 서울CCTV 데이터 중에서 '인구수', '소계' 컬럼을 범죄 정보로 복사
crime_anal_norm[['인구수', 'CCTV']] = result_CCTV[['인구수', '소계']]
crime_anal_norm

# '범죄' 컬럼을 구하는 데 필요한 컬럼:'강간' + '강도' + '살인' + '절도' + '폭력'
col = ['강간', '강도', '살인', '절도', '폭력']

# 5대 중요범죄의 합 계산
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis=1)
crime_anal_norm.head()
```

- 전체적인 상관계수는 CCTV와 살인의 관게가 낮을지 몰라도 CCTV가 없을 때 살인이 많이 일어나는 구간이 있다.  
즉, CCTV 개수를 기준으로 좌측면에 살인과 강도의 높은 수를 갖는 데이터가 보인다.

```python
### 2. ['인구수', 'CCTV'] ['살인', '강도'] : 발생건수
sns.pairplot(crime_anal_norm,
            x_vars=['인구수', 'CCTV'],
            y_vars=['살인검거율', '폭력검거율'],
            kind='reg',
            height=3)
plt.show()
```

> 위에 그래프에서 검거율과 CCTV 상관관계 분석
- 특이한 건 검거율과 CCTV의 상관관계가 양이 아니고 음(-)으로 나타난다.
> 검거율과 인구수의 상관관계
- 음의 상관관계가 보인다

> 검거율에 대한 시각화 heatmap 사용
- 보다 정확한 시각화를 위한 검거율 재계산 
    - 검거율의 합게인 검거 항목 최고 값을 100으로 한정하고 heatmap()을 통한 시각화

```python
# 검거율에 대한 시각화 : heatmap()
col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
# 검거율의 합 => crime_anal_norm
crime_anal_norm['검거'] = np.sum(crime_anal_norm[col2], axis=1)
crime_anal_norm
```

```python
# 검거의 최대값 추출
tmp_max = crime_anal_norm['검거'].max()

# 검거율의 최대 값을 100으로 설정하여 heatmap()을 통한 시각화
# 최대값인 432.5를 100%으로 설정하고 그 비율만큼 재계산
crime_anal_norm['검거'] = crime_anal_norm['검거'] / tmp_max * 100

#전체 범죄의 검거율['검거'] 컬럼을 기분으로 정렬
crime_anal_norm_sort = crime_anal_norm.sort_values(by='검거', ascending=False)
crime_anal_norm_sort.head()
```

```python
# 검거율에 대한 시각화 : heatmap()

plt.figure(figsize=(10,10))
sns.heatmap(crime_anal_norm_sort[col2],
            annot = True, 
            fmt = 'f',
            linewidth=0.5,
            cmap='RdPu')
plt.title('범죄 검거 비율(정규화된 검거의 합으로 계산)')
plt.show()
```

>  ## 범죄 발생 건수의 합으로 계산

- 강남구, 양천구, 영등포구, 송파구 순으로 범죄 발생 빈도가 높다. 서초구도 낮다고 볼수 없다.
- 강남3구는 안전하지 않은 것으로 볼 수 있다.

```python
crime_anal_norm.sort_values(by='범죄', ascending=False)
```

```python
col = ['강간', '강도', '살인', '절도', '폭력', '범죄']

crime_anal_norm['범죄'] = crime_anal_norm['범죄'] / 5
crime_anal_norm_sort = crime_anal_norm.sort_values(by='범죄', ascending=False)

plt.figure(figsize=(10,10))
sns.heatmap(crime_anal_norm_sort[col],
            annot = True, 
            fmt = 'f',
            linewidth=0.5,
            cmap='RdPu')
plt.title('범죄 발생 건수')
plt.show()
```

> ## folium.Choropleth

행정구역 등과 같이 지도상의 어떤 경계로 구분된 지도 영역에 색을 칠하거나 음영 등을 넣어 정보를 나타내는, 단계구분도를 표시하는 기능이다.

```python
import json
import pandas as pd
import folium
import matplotlib.pyplot as plt

# 미국 주별 실업률 데이터 불러오기
state_data = pd.read_csv('./02. folium_US_Unemployment_Oct2012.csv')

# 앱 객체 생성
m = folium.Map([43, -102], zoom_start=3)

# Choropleth 레이어를 만들고, 맵 m에 추가합니다.
folium.choropelth(
    geo_data = './02.folium_us-states.json', # 경계선 좌표값이 담긴 데이터
    data = state_data, #Series or DataFrame 넣은면 된다.
    columns = ['State', 'Unemployment'], # DataFrame의 어떤 컬럼을 넣을지
    key_on = 'feature.id',
    fill_color = 'BuPu',
    fill_opacity = 0.5, # 색 투명도
    line_opacity = 0.5, # 선 투명도
    legend_name = 'Unemployment rate(%)' # 범례
).add_to(m)
m
```
